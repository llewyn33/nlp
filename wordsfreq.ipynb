```python
import pandas as pd
import jieba
import csv
import re
from collections import Counter
from pprint import pprint
import numpy as np

###åˆ‡è¯###

#å¯¼å…¥è‡ªå®šä¹‰è¯å…¸
jieba.load_userdict("userdict.txt")

#å¯¼å…¥åœç”¨è¯è¯å…¸
stopwords = pd.read_csv('stop_words_CN.txt', encoding='utf-8',quoting=csv.QUOTE_NONE,names=['stopword'], index_col=False)
stop_list = stopwords['stopword'].tolist()

#è¯»å–csv/xlsxæ–‡ä»¶
data = pd.read_excel('å’¨è¯¢ç±».xlsx', encoding='utf_8')
# data = data[(data.è¯‰æ±‚åŒºåŸŸ == 'é™å®‰åŒº')]
#data = pd.read_excel('wx.xlsx')
data = data.astype({'å†…å®¹æè¿°':'str'})

#æ–‡æœ¬é¢„å¤„ç†

data.dropna(how = 'any')#æ¸…é™¤ç©ºå€¼
data = data.applymap((lambda x: "".join(x.split()) if type(x) is str else x)) #æ¸…é™¤ç©ºæ ¼


#åˆ‡è¯
data['cut'] = data['å†…å®¹æè¿°'].apply(lambda x : [i for i in jieba.cut(x) if i not in stop_list])
data.head()

###è¯é¢‘ç»Ÿè®¡###

words = []

for content in data['cut']:
    words.extend(content)
    
counter = Counter(words)
output = counter.most_common(300)
output = dict(output)

###ç”Ÿæˆæ–‡ä»¶###

# with open('test_freq.csv', 'w', encoding='utf_8_sig') as f:
#     [f.write('{0},{1}\n'.format(key, value)) for key, value in output.items()]
```
